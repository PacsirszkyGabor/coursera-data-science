### I will import pandas and data from the wikipedia page, then check dataframe's basic features to see if import was successfull.

```python
import pandas as pd
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python
df.shape
```

```python
df.head()
```

### Excluding "Not assigned" boroughs from further analysis, re-checking dataframe

```python
df = df[df.Borough !="Not assigned"]
df.shape
```

```python
df.head()
```

### Grouping data on Postcode level, applying join with "," separator to keep all neighbourhood information. Result saved in a new dataframe.    

```python
df_new = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df_new.shape
```

### Checking if there is any 'Not assigned' neighbourhoods

```python
df_new.loc[df_new['Neighbourhood']=='Not assigned']
```

### Overwriting 'Not assigned' neighbourhood with borough value and checking if it succeded

```python
df_new['Neighbourhood'].loc[df_new['Neighbourhood']=='Not assigned']=df_new['Borough']
df_new.iloc[85:86]
```

<h3> Saving notebook to CSV to be able to work with it in another notebook

```python
df_new.to_csv('df_new.csv')
```
